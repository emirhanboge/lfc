```python
import pandas as pd
import time
import datetime
```

```python
issues = pd.read_pickle("/content/drive/MyDrive/Learning from Contributions/pytorch_issue_data/pickles/issues1")
pullr = pd.read_pickle("/content/drive/MyDrive/Learning from Contributions/pytorch_issue_data/pickles/pullr1")
```

```python
t = []
for idx, row in issues.iterrows():
  if row["state"] == "closed":
    opn = datetime.datetime.strptime(row["open_date"], '%Y-%m-%d %H:%M:%S')
    cls = datetime.datetime.strptime(row["close_date"], '%Y-%m-%d %H:%M:%S')
    diff = (cls - opn)
    diff = diff.total_seconds()
    t.append(diff)
  else:
    t.append(None)

issues["seconds_to_close"] = t

t = []
for idx, row in pullr.iterrows():
  if row["state"] == "closed":
    opn = datetime.datetime.strptime(row["open_date"], '%Y-%m-%d %H:%M:%S')
    cls = datetime.datetime.strptime(row["close_date"], '%Y-%m-%d %H:%M:%S')
    diff = (cls - opn)
    diff = diff.total_seconds()
    t.append(diff)
  else:
    t.append(None)

pullr["seconds_to_close"] = t
```
