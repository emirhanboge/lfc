```python
filenames = [
             "0_1000_issue_body_sentiments.csv",
             "1000_2000_issue_body_sentiments.csv",
             "2000_3000_issue_body_sentiments.csv",
             "3000_4000_issue_body_sentiments.csv",
             "4000_5000_issue_body_sentiments.csv",
             "5000_6000_issue_body_sentiments.csv",
             "6000_7000_issue_body_sentiments.csv",
             "7000_8000_issue_body_sentiments.csv",
             "8000_9000_issue_body_sentiments.csv",
             "9000_10000_issue_body_sentiments.csv",
            "10000_11000_issue_body_sentiments.csv",
             "11000_12000_issue_body_sentiments.csv",
             "12000_13000_issue_body_sentiments.csv",
             "13000_14000_issue_body_sentiments.csv",
             "14000_15000_issue_body_sentiments.csv",
             "15000_16000_issue_body_sentiments.csv",
             "16000_17000_issue_body_sentiments.csv",
             "17000_18000_issue_body_sentiments.csv",
             "18000_19000_issue_body_sentiments.csv",
             "19000_20000_issue_body_sentiments.csv",
             "20000_21000_issue_body_sentiments.csv",
             "21000_22000_issue_body_sentiments.csv",
             "22000_23000_issue_body_sentiments.csv",
             "23000_24000_issue_body_sentiments.csv",
             "24000_24480_issue_body_sentiments.csv",
]
```

```python
label = {}

where = 0
for f in filenames:
  c = 0
  c_ = 0
  for x in f:
    if x == "_" and c_ == 0:
      prev = c+1
      start = int(f[:c])
      c_ = 1
    elif x == "_" and c_ == 1:
      end = int(f[prev:c])
      c_ = 2

    c += 1

  senti = {}
  df = pd.read_csv(f)
  for idx, row in df.iterrows():
    index = int(row["Row"][1:])
    if index != 0:
      senti[index-1+start] = row["Predicted"]

  if (where == len(filenames)-1):
    end += 1

  for i in range(start, end):
    label[i] = senti[i]

  where += 1

issue_body_sentiment = []
for key in label.keys():
  issue_body_sentiment.append(label[key])

issues["issue_body_sentiment"] = issue_body_sentiment
```

```python
neg = 0
pos = 0
neu = 0
for i in issue_body_sentiment:
  if i == "negative":
    neg += 1
  if i == "positive":
    pos += 1
  if i == "neutral":
    neu += 1
```
