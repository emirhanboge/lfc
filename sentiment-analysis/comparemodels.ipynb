```python
def find_label_by_treshold(bert_df):
    labels = []
    for i in range(len(bert_df)):
        pos_prob = bert_df['positive'][i]
        neg_prob = bert_df['negative'][i]
        neu_prob = bert_df['neutral'][i]

        if pos_prob >= 0.7:
            labels.append('positive')
        elif neg_prob >= 0.7:
            labels.append('negative')
        else:
            labels.append('neutral')
    return labels
```

```python
# compare the results of the two models
import pprint as pp
count = 0
d = {'positive': 0, 'negative': 0, 'neutral':0}
index = 0
#labels = find_label_by_treshold(bert)
for idx, row in sentisd.iterrows():
    bert_prediction = bert['label'][idx]
    senti4sd_prediction = row['issue_body_sentiment']

    if senti4sd_prediction == bert_prediction:
        count += 1
        d[bert_prediction] += 1

    if bert_prediction == "negative" and index < 5:
      pp.pprint(row["body"])
      print()
      print()

      index += 1

count/len(sentisd) * 100
```
