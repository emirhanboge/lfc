### Delete Codeblocks

```python
body = []
for idx, row in issues.iterrows():
  code = 0
  text = str(row["body"])
  new = ""
  for i in range(0, len(text)): 
    if i!=0 and i!= len(text)-1:
      if text[i] == "`" and (text[i-1] == "`" or text[i+1] == "`"):
        code += 1
    elif i != 0:
      if text[i] == "`"  and text[i-1] == "`":
        code += 1
    elif i != len(text)-1: 
      if text[i] == "`" and text[i+1] == "`":
        code += 1

    if not code > 0:
      new += text[i]
    if code == 6:
      code = 0
  body.append(new)

# remove \r and unnecessary  whitespaces
for i in range(0, len(body)):
  body[i] = body[i].replace('\r', "")
  body[i] = body[i].strip()

comments = []
for idx, row in issues.iterrows():
  l = []
  for x in row["comments"]:
    code = 0
    text = x
    new = ""
    for i in range(0, len(text)): 
      if i!=0 and i!= len(text)-1:
        if text[i] == "`" and (text[i-1] == "`" or text[i+1] == "`"):
          code += 1
      elif i != 0:
        if text[i] == "`"  and text[i-1] == "`":
          code += 1
      elif i != len(text)-1: 
        if text[i] == "`" and text[i+1] == "`":
          code += 1

      if not code > 0:
        new += text[i]
      if code == 6:
        code = 0
    l.append(new)
  comments.append(l)

# remove \r and unnecessary  whitespaces
for x in range(0,len(comments)):
  for i in range(0, len(comments[x])):
    comments[x][i] = comments[x][i].replace('\r', "")
    comments[x][i] = comments[x][i].strip()

issues["body"] = body
issues["comments"] = comments
```

#### Delete Codes from Texts

```python
body = []
for idx, row in issues.iterrows():
  sentences = []
  for x in row["body"].split("\n"):
    if x != "" and x!=" ":
      sentences.append(x)

  clean_sentences = []
  for text in sentences:
    new = []
    sent_text = nltk.sent_tokenize(text)
    for i in sent_text:
      s = ""
      for x in i.split():
        if x.lower() in contractions:
          x = contractions[x.lower()]
        s += x + " "

      tokenized_text = nltk.word_tokenize(s)
      for x in tokenized_text:
        if x == "." or x == "!" or x == "?":
          new.append(x)
        elif x.isalpha() and x.lower():
          new.append(x.lower())
    s = ""
    for x in range(0, len(new)):
      if x == 0 or new[x] == "." or new[x] == "!" or new[x] == "?":
        s += new[x]
      else:
        s += " " + new[x]
    clean_sentences.append(s)

  s = ""
  for x in clean_sentences:
    last = "."
    if x != "":
      last = x[-1]
    if last != "." and last != "?" and last != "!":
      x += "."
    s += x + " "

  s = s.strip()
  body.append(s)

comments = []
for idx, row in issues.iterrows():
  l = []
  for k in row["comments"]:
    sentences = []
    for x in k.split("\n"):
      if x != "" and x!=" ":
        sentences.append(x)

    clean_sentences = []
    for text in sentences:
      new = []
      sent_text = nltk.sent_tokenize(text)
      for i in sent_text:
        s = ""
        for x in i.split():
          if x.lower() in contractions:
            x = contractions[x.lower()]
          s += x + " "

        tokenized_text = nltk.word_tokenize(s)
        for x in tokenized_text:
          if x == "." or x == "!" or x == "?":
            new.append(x)
          elif x.isalpha() and x.lower():
            new.append(x.lower())
      s = ""
      for x in range(0, len(new)):
        if x == 0 or new[x] == "." or new[x] == "!" or new[x] == "?":
          s += new[x]
        else:
          s += " " + new[x]
      clean_sentences.append(s)
    s = ""
    for x in clean_sentences:
      last = "."
      if x != "":
        last = x[-1]
      if last != "." and last != "?" and last != "!":
        x += "."
      s += x + " "

    s = s.strip()
    l.append(s)
  comments.append(l)

issues["body"] = body
issues["comments"] = comments
```

### tf-idf

```python
# Find tf-idf for each issue and pull request
from sklearn.feature_extraction.text import TfidfVectorizer
import string

vectorizer = TfidfVectorizer()

# Create a list of all the sentences in the issues and pull requests data and remove punctuation
all_sentences = []
for idx, row in issues.iterrows():
  # Remove punctuation
  s = row["body"]
  s = s.translate(str.maketrans('', '', string.punctuation))
  all_sentences.append(s)
  for k in row["comments"]:
    s = k
    s = s.translate(str.maketrans('', '', string.punctuation))
    all_sentences.append(s)

for idx, row in pullr.iterrows():
  # Remove punctuation
  s = row["body"]
  s = s.translate(str.maketrans('', '', string.punctuation))
  all_sentences.append(s)
  for k in row["comments"]:
    s = k
    s = s.translate(str.maketrans('', '', string.punctuation))
    all_sentences.append(s)

# Fit the vectorizer to the data
vectorizer.fit(all_sentences)

# Print vocabulary length
print(len(vectorizer.vocabulary_))
```

```python
tfidf_matrix = vectorizer.fit_transform(all_sentences)

# Find the cosine similarity between the TF-IDF matrix of the issue and pull request bodies
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix[0:len(issues)], tfidf_matrix[len(issues):])
```
