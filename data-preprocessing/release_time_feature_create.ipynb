```python
for key in releases.keys():
    releases[key] = time.strptime(releases[key], '%b %d, %Y')
releases["v0.1.1"]
```

```python
release = []
for index, row in issues.iterrows():
    opn = time.strptime(row["open_date"][:10], '%Y-%m-%d')
    prev_key = ""
    found = False
    for key in releases.keys():
        if releases[key] >= opn > releases[prev_key]:
            release.append(key)
            found = True
        prev_key = key
        
    if not found:
        print(index)
        break
```

```python
issues['release'] = release
for idx, row in issues.iterrows():
    if row["release"] == "v0.1.8":
        row["release"] = "v0.1.9"
```
