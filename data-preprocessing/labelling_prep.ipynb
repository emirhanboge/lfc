```python
import pandas as pd
```

```python
issues = pd.read_pickle("/content/drive/MyDrive/Learning from Contributions/pytorch_issue_data/pickles/cleanissues")
pullr = pd.read_pickle("/content/drive/MyDrive/Learning from Contributions/pytorch_issue_data/pickles/cleanpullr")
```

```python
issue_comments = []
issue_body = []
count = 0
for idx, row in issues.iterrows():
    count += len(row['comments']) + 1
    issue_comments.append(row['comments'])
    issue_body.append(row['body'])

pull_comments = []
pull_body = []
for idx, row in pullr.iterrows():
    count += len(row['comments']) + 1
    pull_comments.append(row['comments'])
    pull_body.append(row['body'])
```

```python
count
```

```python
# save the data to a dataframe
df_issues = pd.DataFrame({'issue_comments': issue_comments, 'issue_body': issue_body})
df_pulls = pd.DataFrame({'pull_comments': pull_comments, 'pull_body': pull_body})
```

```python
# save the dataframes to csvs
df_issues.to_csv("/content/drive/MyDrive/Learning from Contributions/labelling/issues/issuelabels.csv")
df_pulls.to_csv("/content/drive/MyDrive/Learning from Contributions/labelling/pull_requests/pullrlabels.csv")
```
