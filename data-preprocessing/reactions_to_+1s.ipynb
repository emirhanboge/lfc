```python
import pandas as pd
```

```python
issues = pd.read_pickle("/content/drive/MyDrive/Learning from Contributions/pytorch_issue_data/pickles/issues")
pullr = pd.read_pickle("/content/drive/MyDrive/Learning from Contributions/pytorch_issue_data/pickles/pullr")
```

```python
react=[]
for idx, row in issues.iterrows():
  x = row["reactions"]["+1"]
  react.append(x)

issues["reactions"] = react

react=[]
for idx, row in pullr.iterrows():
  x = row["reactions"]["+1"]
  react.append(x)

pullr["reactions"] = react
```

```python
issues = issues.rename(columns={"reactions":"likes"})
pullr = pullr.rename(columns={"reactions":"likes"})
```
